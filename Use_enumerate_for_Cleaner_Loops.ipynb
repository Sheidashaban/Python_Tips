```python
# ðŸ§  Python Tip: Use enumerate() instead of range(len(...))

# Let's say we have a list of fruits
fruits = ["apple", "banana", "cherry", "mango"]

print("ðŸ”´ Traditional way using range(len(...)):")
for i in range(len(fruits)):
    print(f"Index: {i}, Fruit: {fruits[i]}")

print("\nðŸŸ¢ Pythonic way using enumerate():")
for i, fruit in enumerate(fruits):
    print(f"Index: {i}, Fruit: {fruit}")

print("\nðŸŸ£ Starting index from 1 using enumerate(..., start=1):")
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")

# Bonus: Using enumerate() with a conditional example
print("\nðŸ’¡ Example: Find the index of 'cherry'")
for i, fruit in enumerate(fruits):
    if fruit == "cherry":
        print(f"Found '{fruit}' at index {i}")
```
