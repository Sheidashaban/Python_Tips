# Python Tip: Using f-strings for formatting

F-strings (formatted string literals) provide a concise and readable way to embed expressions inside string literals. They're faster and more readable than older formatting methods.

**Generated on:** 2025-11-02

```python
name = "Alice"age = 30city = "Paris"# Old way with %message = "My name is %s, I'm %d years old, from %s" % (name, age, city)# Better with .format()message = "My name is {}, I'm {} years old, from {}".format(name, age, city)# Best with f-strings (Python 3.6+)message = f"My name is {name}, I'm {age} years old, from {city}"print(message)# F-strings can include expressionsprint(f"Next year I'll be {age + 1} years old")
```
