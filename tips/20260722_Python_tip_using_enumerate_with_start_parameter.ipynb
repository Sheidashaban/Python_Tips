# Python Tip: Using enumerate with start parameter

The start parameter in enumerate() allows you to begin counting from any number, making it easy to create numbered lists starting from 1 or any other value.

**Generated on:** 2026-07-22

```python
tasks = ['Write code', 'Review PR', 'Deploy', 'Monitor']# Default - starts from 0for i, task in enumerate(tasks):    print(f"{i}. {task}")print()# Start from 1 for human-readable numberingfor i, task in enumerate(tasks, start=1):    print(f"{i}. {task}")# Start from any numberfor i, task in enumerate(tasks, start=100):    print(f"Task #{i}: {task}")
```
