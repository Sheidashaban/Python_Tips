# Python Tip: Avoid Mutating a List While Iterating

Modifying a list while iterating over it can lead to unexpected behavior. Create a copy of the list to iterate over to avoid issues.

**Generated on:** 2025-12-19

````python
```pythonoriginal_list = [1, 2, 3, 4, 5]for num in original_list[:]:  # Iterate over a copy of the original list    if num % 2 == 0:        original_list.remove(num)  # Remove even numbers from the original listprint(original_list)  # Output: [1, 3, 5]```COMMENTS:- By iterating over a copy of `original_list` using `original_list[:]`, we avoid modifying the list we are iterating over.- Modifying a list directly while iterating can change its length, affecting the loop behavior and potentially causing items to be skipped.
````
