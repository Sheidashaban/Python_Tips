# Python Tip: Context managers with statement

Use context managers (with statement) to ensure resources are properly managed and cleaned up, even if exceptions occur. Most commonly used with file operations.

**Generated on:** 2025-11-02

```python
# Without context manager (bad practice)file = open('example.txt', 'r')data = file.read()file.close()  # What if an error occurs before this?# With context manager (recommended)with open('example.txt', 'r') as file:    data = file.read()    # File automatically closes when leaving the block# Multiple context managerswith open('input.txt', 'r') as infile, open('output.txt', 'w') as outfile:    for line in infile:        outfile.write(line.upper())
```
