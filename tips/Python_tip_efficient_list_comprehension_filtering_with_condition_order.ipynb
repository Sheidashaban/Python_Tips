# Python Tip: Efficient List Comprehension Filtering with Condition Order

Placing more specific conditions earlier can improve performance in list comprehensions.

**Generated on:** 2025-12-27

````python
```pythonnumbers = [1, 2, 3, 4, 5, 6, 7, 8, 9]# Filter even numbers first, then multiples of 3filtered_numbers = [num for num in numbers if num % 2 == 0 if num % 3 == 0]print(filtered_numbers)  # Output: [6]```COMMENTS:- In this example, filtering even numbers before multiples of 3 can reduce the number of checks performed.- Placing the more specific condition first optimizes the filtering process.
````
