# Python Tip: Efficiently Check if All Elements in a List are True

Use the `all()` function with a list comprehension to efficiently check if all elements in a list satisfy a condition.

**Generated on:** 2025-12-22

````python
```python# Check if all elements in a list are even numbersnums = [2, 4, 6, 8]all_even = all(num % 2 == 0 for num in nums)print(all_even)  # Output: True```COMMENTS:- The list comprehension `num % 2 == 0 for num in nums` generates a list of boolean values based on the condition `num % 2 == 0`.- The `all()` function checks if all elements in the list are `True`.
````
