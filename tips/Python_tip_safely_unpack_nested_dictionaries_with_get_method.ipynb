# Python Tip: Safely Unpack Nested Dictionaries with get() Method

Avoid KeyError by safely unpacking nested dictionaries using Python's get() method.

**Generated on:** 2025-11-15

````python
```pythondata = {    'user': {        'name': 'Alice',        'age': 30    }}# Safely unpack nested dictionaryname = data.get('user', {}).get('name')print(name)  # Output: Alice```COMMENTS:The `get()` method returns the value for a given key, or a default value (in this case an empty dictionary {}) if the key is not found. This prevents raising a KeyError when accessing nested dictionary keys.
````
