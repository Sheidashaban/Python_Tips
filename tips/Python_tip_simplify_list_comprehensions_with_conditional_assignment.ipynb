# Python Tip: Simplify List Comprehensions with Conditional Assignment

Use conditional assignment in list comprehensions for clearer and more concise code.

**Generated on:** 2025-12-20

````python
```python# Example: Create a list of squared numbers only for positive numbers in the original listoriginal_list = [1, -2, 3, -4, 5]squared_positives = [num**2 for num in original_list if num > 0]print(squared_positives)  # Output: [1, 9, 25]```# In this example, the list comprehension only squares the numbers that are greater than 0, simplifying the logic and making the code more readable.
````
