# Python Tip: Using collections.defaultdict for Default Values in Dictionaries

defaultdict from the collections module allows setting default values to keys that do not exist in a dictionary.

**Generated on:** 2025-12-24

````python
```pythonfrom collections import defaultdict# Create a defaultdict with int as default valuescores = defaultdict(int)# Access and update scores without worrying about key existencescores['Alice'] += 1scores['Bob'] += 2print(scores)  # Output: defaultdict(<class 'int'>, {'Alice': 1, 'Bob': 2})```COMMENTS:- The defaultdict(int) constructor sets the default value type to int.- When accessing a non-existent key in scores and performing operations, the default value (0 for int) is used without raising a KeyError.
````
