# Python Tip: Using Counter for counting hashable objects

The Counter class from collections is designed specifically for counting occurrences of elements, making it much cleaner than manual dictionary tracking.

**Generated on:** 2026-06-30

```python
from collections import Counter# Count occurrences in a listwords = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']word_counts = Counter(words)print(word_counts)  # Counter({'apple': 3, 'banana': 2, 'cherry': 1})# Get most common elementsprint(word_counts.most_common(2))  # [('apple', 3), ('banana', 2)]# Count characters in a stringtext = "hello world"char_counts = Counter(text)print(char_counts)  # Counter({'l': 3, 'o': 2, 'h': 1, ...})# Combine countersmore_words = Counter(['apple', 'date'])total_counts = word_counts + more_wordsprint(total_counts)
```
