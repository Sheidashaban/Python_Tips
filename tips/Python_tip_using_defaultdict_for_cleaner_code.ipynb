# Python Tip: Using defaultdict for cleaner code

defaultdict from collections automatically initializes missing keys with a default value, eliminating the need for key existence checks.

**Generated on:** 2026-07-02

```python
from collections import defaultdict# Regular dict - need to check if key existsword_count = {}words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']for word in words:    if word not in word_count:        word_count[word] = 0    word_count[word] += 1# defaultdict - cleaner approachword_count = defaultdict(int)for word in words:    word_count[word] += 1print(dict(word_count))  # {'apple': 3, 'banana': 2, 'cherry': 1}
```
