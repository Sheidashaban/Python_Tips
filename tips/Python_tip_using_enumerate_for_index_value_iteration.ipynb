# Python Tip: Using Enumerate() for Index-Value Iteration

Enumerate() simplifies iteration by providing index-value pairs.

**Generated on:** 2025-11-14

````python
```pythonfruits = ['apple', 'banana', 'cherry']for index, fruit in enumerate(fruits):    print(f"Index {index}: {fruit}")```COMMENTS:- Enumerate() returns both the index and value of each item in the iterable.- This is useful when you need to access both the index and value in a loop.
````
