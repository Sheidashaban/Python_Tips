# Python Tip: Using enumerate() for Index-Value Pair Iteration

The enumerate() function in Python allows you to iterate over both the index and value of elements in a sequence simultaneously, enhancing code readability and efficiency.

**Generated on:** 2025-12-18

````python
```pythonfruits = ['apple', 'banana', 'cherry']# Enumerate over the list of fruits to print index and value pairfor index, fruit in enumerate(fruits):    print(f'Index: {index}, Value: {fruit}')```COMMENTS:- The enumerate() function returns a tuple containing the index and value of each element in the iterable.- In the example, we iterate over a list of fruits and print the index and value pair for each element.
````
