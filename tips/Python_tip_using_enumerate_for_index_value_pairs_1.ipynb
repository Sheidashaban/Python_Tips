# Python Tip: Using Enumerate for Index-Value Pairs

Enumerate allows iterating over a collection with both index and value.

**Generated on:** 2025-12-23

````python
```pythonmy_list = ['apple', 'banana', 'cherry']for index, value in enumerate(my_list):    print(f'Index: {index}, Value: {value}')```COMMENTS:- `enumerate()` returns index-value pairs, making it useful for iterating over lists while also keeping track of the index.- In this example, the loop prints the index and value of each element in the list `my_list`.
````
