# Python Tip: Using Enumerate for Index-Value Pairs

Enumerate simplifies iterating over a sequence with index-value pairs.

**Generated on:** 2025-11-08

````python
```pythonfruits = ['apple', 'banana', 'cherry']for index, value in enumerate(fruits):    print(f'Index: {index}, Value: {value}')```COMMENTS:- `enumerate()` adds a counter to an iterable and returns it.- The loop iterates over the index-value pairs of the `fruits` list.
````
