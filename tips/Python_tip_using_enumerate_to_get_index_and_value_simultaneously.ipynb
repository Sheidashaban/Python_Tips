# Python Tip: Using Enumerate to Get Index and Value Simultaneously

Enumerate simplifies iterating by providing both index and value.

**Generated on:** 2025-11-20

````python
```pythonfruits = ['apple', 'banana', 'cherry']for index, value in enumerate(fruits):    print(f'Index: {index}, Value: {value}')```COMMENTS:- The `enumerate()` function returns both the index and value of each item in the iterable.- This allows you to access the index and value together without needing to manually manage indices.
````
