# Python Tip: Using enumerate() with start parameter for custom index

Use enumerate() with a start parameter to specify a custom starting index for iteration, useful when you need to control the index values in your loop.

**Generated on:** 2025-12-13

````python
```python# Enumerate with custom starting indexfruits = ['apple', 'banana', 'cherry']for idx, fruit in enumerate(fruits, start=1):  # Start index from 1    print(f'{idx}. {fruit}')```COMMENTS:- In this example, enumerate() is used with the start parameter set to 1 to start the index from 1 instead of the default 0.- This can be useful when you want a more human-readable index, such as for displaying numbered lists to users.
````
