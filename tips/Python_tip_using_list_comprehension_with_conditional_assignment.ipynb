# Python Tip: Using List Comprehension with Conditional Assignment

Utilize list comprehension to conditionally assign values based on a condition.

**Generated on:** 2025-12-28

````python
```pythonnumbers = [1, 2, 3, 4, 5]squared_numbers = [num**2 if num % 2 == 0 else num for num in numbers]print(squared_numbers)  # Output: [1, 4, 3, 16, 5]```COMMENTS:- In this code, we square the number if it's even (num % 2 == 0) using a conditional assignment in list comprehension.- This technique allows for concise and readable code to transform elements in a list based on conditions.
````
