# Python Tip: Using namedtuple for Immutable Data Structures

namedtuple from the collections module creates lightweight, immutable data structures with named fields, enhancing readability and avoiding accidental mutation.

**Generated on:** 2025-11-11

````python
```pythonfrom collections import namedtuple# Define a namedtuple 'Point' with fields x and yPoint = namedtuple('Point', ['x', 'y'])# Create an instance of Pointp = Point(3, 5)# Access fields by nameprint(p.x, p.y)```COMMENTS:- The `namedtuple` function is used to create a new class with named fields, offering readability benefits over tuples.- Immutable nature of `namedtuple` instances prevents accidental changes, enhancing code safety.
````
