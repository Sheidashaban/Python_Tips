# Python Tip: Using pathlib for file operations

The pathlib module provides an object-oriented way to work with file paths, making code more readable and cross-platform compatible.

**Generated on:** 2026-07-01

```python
from pathlib import Path# Create path objectsproject_dir = Path('my_project')data_file = project_dir / 'data' / 'input.txt'# Check if file existsif data_file.exists():    print(f"File found: {data_file}")# Read file contentcontent = data_file.read_text()# Get file propertiesprint(f"File size: {data_file.stat().st_size} bytes")print(f"File name: {data_file.name}")print(f"Parent directory: {data_file.parent}")
```
