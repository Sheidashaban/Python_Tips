# Python Tip: Using the `any` Function for List Comprehensions

The `any` function can be used in list comprehensions to check if at least one element satisfies a condition. It provides a concise way to filter elements in a list based on a condition.

**Generated on:** 2025-12-25

````python
```python# Check if any element in the list is divisible by 5numbers = [12, 15, 18, 21, 25, 30]divisible_by_5 = [num for num in numbers if any(num % 5 == 0 for num in numbers)]print(divisible_by_5)  # Output: [15, 25, 30]```# The `any` function is used inside the list comprehension to check if any number in the list is divisible by 5. If the condition `num % 5 == 0` is True for any element, that element is included in the new list `divisible_by_5`.
````
