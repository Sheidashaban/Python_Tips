# Python Tip: Utilize Enumerate for Index-Value Pairs

Enumerate simplifies looping by providing index-value pairs in iterations.

**Generated on:** 2025-11-16

````python
```pythonfruits = ['apple', 'banana', 'cherry']for index, fruit in enumerate(fruits):    print(f'Index: {index}, Value: {fruit}')```COMMENTS:- The `enumerate()` function is used to iterate over both the index and value in the `fruits` list.- This is especially useful when you need to track the index along with the value in a loop.
````
